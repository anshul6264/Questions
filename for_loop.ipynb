```python
# 1. Write a program to print numbers from 1 to 10 using for loop.

for i in range(1,11):
    print(i,end=" ")
```

```python
# 2. Write a program using a for loop to print all even numbers from 1 to 20.

for i in range(2,21,2):
    print(i,end=" ")
```

```python
# 3. Write a program to find the sum of numbers from 1 to 10 using a loop.

sum=0
for i in range(1,11):
    sum=sum+i
print(sum)
```

```python
# 4. Write a program that prints the numbers from 10 to 1 using a for loop. 

for i in range(10,0,-1):
    print(i,end=" ")
```

```python
# 5. Write a program that print the square of each number from 1 to 10.

import math
for i  in range(1,11):
    print(f"square of {i} is ",math.pow(i,2))
```

```python
# 6. Write a program to display the multiplication table of a number entered by the user.

num=int(input("enter a number :"))
for i in range(1,11):
    mul=num*i
    print(f"{num} x {i} = {mul} ")
```

```python
# 7. Print all numbers between 1–500 that are divisible by 7 but not by 5

for i in range(1,501):
    if i%7==0 and not i%5==0:
        print(i,end=" ")
```

```python
# 8. Find all prime numbers between 1 and 100

for i in range(2,101):
    for j in range(2,i):
        if i%j==0:
            break
    else:
        print(i,end=" ")
```

```python
# 9. Check whether a number is a perfect square

num=int(input("enter a number: "))
for i in range(1,num+1):
    if i*i==num:
        print(f"{num} is the perfect square of {i}")
        break
else:
    print(f"{num} is not a perfect square...")
```

```python
# 10. Print pattern using loop
#     *
#     **
#     ***
#     ****

for i in range(1,5):
    print(i*"*")
```
