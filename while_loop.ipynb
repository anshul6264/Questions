```python
# 1. Print numbers from 1 to 10 using a while loop.

i=1
while i<=10:
    print(i,end=" ")
    i+=1
```

```python
# 2. Print numbers from 10 to 1 using a while loop.

i=10
while i>=1:
    print(i,end=" ")
    i-=1
```

```python
# 20. Implement FizzBuzz using a while loop.

i=1
while i<=100:
    if i%3==0 and i%5==0:
        print("Fizzbuzz",end=" ")
    elif i%5==0:
        print("Buzz",end=" ")
    elif i%3==0:
        print("Fizz",end=" ")
    else:
        print(i,end=" ")
    i+=1
```
