```python
# generate 5 random nos. find evn odd and count how many even and odd are there (1,50)

import random
# help(random.randint)
# help(random.random)
# help(random.randrange)

even=0
odd=0
for i in range (5) :
    a=(random.randint(1,50))
    print(a,end=" ")
    if a%2==0:
        even=even+1
        print("is even")
    else:
        odd=odd+1
        print("is odd")
print("even count is :",even)
print("odd count is :",odd)
        
    

```
