```python
# 4 Nested if and if-elif-else : Largest of Three Numbers

a=int(input("enter 1st number :"))
b=int(input("enter 2nd number :"))
c=int(input("enter 3rd number :"))

if a>b and a>c:
    print(a,"is largest")

elif b>a and b>c:
    print(b,"is largest")
    
else:
    print(c,"is largest")
```
